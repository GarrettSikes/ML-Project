```python
price_data = pd.read_csv('price_df_floor_check.csv')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(price_data['SalePrice'])
plt.show()
```

```python
correlation_matrix = price_data.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
plt.figure(figsize=(20, 5))

features = ['TotRmsAbvGrd', 'GrLivArea', 'OverallQual', 'YearBuilt', 'GarageArea']
target = price_data['SalePrice']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = price_data[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('SalePrice')
    print(x.corr(y))
```

```python
plt.figure(figsize=(20, 5))

features = ['finished_bsmt_sf_total', 'TotalBsmtSF']
target = price_data['OverallQual']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = price_data[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('OverallQual')
    print(x.corr(y))
```
